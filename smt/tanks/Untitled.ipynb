```python
import numpy as np

from random import randrange as rnd, choice
```

```python
file = open('rezults.csv', 'r')
rezults = file.read()
file.close()
```

```python
ar = []
for i in rezults.split('\n'):
    ar.append(i.split(','))
ar.pop(-1)
print(ar)
```

```python
z = []
for i in range(len(ar)):
    z.append((ar[i][0], ar[i][1]))
dtype = [('number', int), ('points', int)]
rez = np.array(z, dtype = dtype)
rez = np.sort(rez, order = 'points')
print(rez)
```

```python
rez = rez[25:50]
print(rez)
```

```python
winners = []
for i in rez:
    winners.append(i[0])
winners.sort()
print(winners)
```

```python
file = open('tanks_weights.csv', 'r')
weights = file.read()
file.close()
ar = []
for i in weights.split('\n'):
    ar.append(i.split(','))
ar.pop(-1)
print(ar)
```

```python
new_weights = []
for i in winners:
    new_weights.append(ar[i]) 
print(new_weights)
```

```python
def crossingover(a, b):
    c = []
    for i in range(len(a)):
        r = rnd(0, 2)
        if r == 0:
            c.append(a[i])
        else:
            c.append(b[i])
    return c
```

```python
for i in range(25):
    new_weights.append(crossingover(new_weights[i],new_weights[i + 1]))
```

```python
file = open('tanks_weights.csv', 'w')
for i in range(len(new_weights)):
    for j in range(len(new_weights[i])):
        file.write(new_weights[i][j])
        if j != len(new_weights[i]) - 1:
            file.write(',')
    file.write('\n')
file.close()
```
